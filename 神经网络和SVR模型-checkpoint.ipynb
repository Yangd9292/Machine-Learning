```python
from keras.preprocessing import sequence  
from keras.models import Sequential   
from keras.layers import Dense, Dropout  
from keras.utils import multi_gpu_model  
from keras import regularizers  # 正则化  
import matplotlib.pyplot as plt  
import numpy as np  
from sklearn.preprocessing import MinMaxScaler  
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split#加载数据
 
x_train, x_test, y_train, y_test = train_test_split(X,Y['pIC50'], test_size=0.1, random_state=29)

# 转成DataFrame格式方便数据处理  
x_train_pd = pd.DataFrame(x_train)  
y_train_pd = pd.DataFrame(y_train)  
x_test_pd = pd.DataFrame(x_test)  
y_test_pd = pd.DataFrame(y_test)  
print(x_train_pd.head(5))  
print('-------------------')  
print(y_train_pd.head(5))
```

```python
#数据归一化
min_max_scaler = MinMaxScaler()  
min_max_scaler.fit(x_train_pd)  
x_train_g= min_max_scaler.transform(x_train_pd)  
  
min_max_scaler.fit(y_train_pd)  
y_train_g = min_max_scaler.transform(y_train_pd)  
   
# 验证集归一化  
min_max_scaler.fit(x_test_pd)  
x_test_g = min_max_scaler.transform(x_test_pd)  
min_max_scaler.fit(y_test_pd)  
y_test_g = min_max_scaler.transform(y_test_pd)
#print(x_train_g[0:5])
#print('-------------------')  
print(y_train_g[0:5])
```

```python
# 径向基核函数配置支持向量机
rbf_svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
# 训练
rbf_svr.fit(x_train, y_train)
# 预测 保存预测结果
rbf_svr_y_predict = rbf_svr.predict(x_test)
plt.plot(y_test.reshape(1,-1)[0])
plt.plot(rbf_svr_y_predict.reshape(1,-1)[0])
plt.show()
```

```python
#训练模型

model = Sequential()  # 初始化，很重要！
model.add(Dense(units = 10,   # 输出大小  
                activation='relu',  # 激励函数  
                input_shape=(x_train_pd.shape[1],)  # 输入大小, 也就是列的大小  
                 )  
           )    
model.add(Dropout(0.2))  # 丢弃神经元链接概率  
    
model.add(Dense(units = 15,  
#kernel_regularizer=regularizers.l2(0.01),  # 施加在权重上的正则项  
#ctivity_regularizer=regularizers.l1(0.01),  # 施加在输出上的正则项  
activation='relu'# 激励函数  
#bias_regularizer=kears.regularizers.l1_l2(0.01)  # 施加在偏置向量上的正则项  
                )  
          )    
model.add(Dense(units = 1,     
                activation='linear'  # 线性激励函数 回归一般在输出层用这个激励函数    
                )  
          )  
    
print(model.summary())  # 打印网络层次结构  
    
model.compile(loss='mse',  # 损失均方误差  
              optimizer='adam',  # 优化器  
              )  
history = model.fit(x_train_g, y_train_g,  
            epochs=300,  # 迭代次数  
            batch_size=200,  # 每次用来梯度下降的批处理数据大小  
            verbose=2,  # verbose：日志冗长度，int：冗长度，0：不输出训练过程，1：输出训练进度，2：输出每一个epoch  
            validation_data = (x_test_g, y_test_g)  # 验证集  
          )
```

```python
import matplotlib.pyplot as plt  
# 绘制训练 & 验证的损失值  
plt.plot(history.history['loss'])  
plt.plot(history.history['val_loss'])  
plt.title('Model loss')  
plt.ylabel('Loss')  
plt.xlabel('Epoch')  
plt.legend(['Train', 'Test'], loc='upper left')  
plt.show()
```
